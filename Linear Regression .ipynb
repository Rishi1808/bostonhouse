```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# URL for the Boston Housing dataset
data_url = "http://lib.stat.cmu.edu/datasets/boston"

# Reading the raw dataset, skipping the first 22 rows (metadata) and using regular expression for space separation
raw_df = pd.read_csv(data_url, sep=r"\s+", skiprows=22, header=None)

# Prepare the feature data and target
data = py.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])  # Combining two parts of the dataset
target = raw_df.values[1::2, 2]  # Extracting target values from the dataset

print("Data shape:", data.shape)
print("Target shape:", target.shape)
```

```python
X=dataset.iloc[:,:-2]
Y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
## Train Test Split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
## Standardizind dataset                               #interllly we use gredient decent and global minima  , to ensure all independet in same scale
## Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler, open('scaler.pkl', 'wb'))
```

```python
X_train
```

```python
## Model Training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## Prediction on test data
```

```python
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
## plot a scatter from prediction
```

```python
plt.scatter(y_test,reg_pred)
plt.show()
```

```python
## Prediction with residuals
```

```python
residuals=y_test-reg_pred
```

```python
residuals
```

```python
## plot this residuals
```

```python
sns.displot(residuals,kind="kde")
plt.show()
```

```python
## scatter plot with respect tot residuals
plt.scatter(reg_pred,residuals)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(py.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
